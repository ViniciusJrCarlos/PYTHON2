```python
import numpy as np
import matplotlib.pyplot as plt


data1 = [10,5,2,4,6,8]
data2 = [ 1,2,4,8,7,4]
x = 10*np.array(range(len(data1)))

plt.plot( x, data1, 'go') # green bolinha
plt.plot( x, data1, 'k:', color='orange') # linha pontilha orange

plt.plot( x, data2, 'r^') # red triangulo
plt.plot( x, data2, 'k--', color='blue')  # linha tracejada azul

plt.axis([-10, 60, 0, 11])
plt.title("Gr√°fico de Linhas")

plt.grid(True)
plt.xlabel("eixo horizontal")
plt.ylabel("eixo vertical")
plt.show()
```

```python
import pandas as pd
import html5lib
import os
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import Request, urlopen
import numpy as np 

req = Request('https://www.portalbrasil.net/ipca/', headers={'User-Agent': 'Mozilla/5.0'})
pagina = urlopen(req).read()
ipca = pd.read_html(pagina)
ipca = ipca[1]
print(ipca)
ipca.head()

ipca.head(10)
ipca.columns

ipca.columns = ['mesano', 'indiceporcentagemmes', 'indiceporcentagemano',
       'indiceultimo12meses',
       'indiceacumuladodesde93']

ipca100 = ipca[0:20]	 


sns.set_style('darkgrid')
plt.title('IPCA Mensal')
plt.legend(loc=7)
plt.xlabel('M√™s/ano')
plt.ylabel('√çndice acumulado nos √∫ltimos 12 meses (em %)')

plt.bar(ipca100.mesano, ipca100.indiceporcentagemmes, label="IPCA", color = "green")
plt.rcParams['figure.figsize'] = (20,11)
plt.rcParams.update({'font.size': 12})
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Visualiza√ß√£o simples")
plt.rcParams.update({'font.size': 20})
plt.show()
```
